# Ch 6 - Loops

## for and while

- `while` = gets a boolian condition that decides if it's running or not
-   between 0 to inifinity

- `for` = gets a list, or finite number of iteration - and runs on another element in the list that we can use

```python
i = 0
while i < 10:
    print(i)
    i = i + 1   
```

```python
names = ['Bob', 'Alice', 'Eve', 'Adam']
names
```

```python
for name in names:
    print(f"hello {name}")
```

```python
for i in range(5):
    print(f"current i = {i}")
```
